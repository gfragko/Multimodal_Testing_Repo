```python
from transformers import Qwen2VLForConditionalGeneration, AutoTokenizer, AutoProcessor
from qwen_vl_utils import process_vision_info


# default: Load the model on the available device(s)
model = Qwen2VLForConditionalGeneration.from_pretrained(
    "Qwen/Qwen2-VL-7B-Instruct", torch_dtype="auto", device_map="auto"
)
```

```python
processor = AutoProcessor.from_pretrained("Qwen/Qwen2-VL-7B-Instruct")
```

## **Pear Tree Analysis**
mature_tree.jpg

```python
from PIL import Image
image = Image.open("/content/mature_tree.jpg")
# Resize the image, for example, to half its original size
image = image.resize((image.width // 2, image.height // 2))
TREE_PROMPT = """You are an agriculture expert, analyze the provided image and answer the following questions about the tree depicted:

1) Tree Identification: What type of tree is this (e.g., pear tree, apple tree, etc.)?
2) Harvest Information: Based on the tree type and visible fruit maturity, when is the approximate harvest time (in how many weeks)?
3) Health Status: Does the tree show any visible signs of disease or distress (e.g., discoloration, spots, or abnormal growth)?
4) Water Needs: Based on the condition of the leaves, soil, and overall tree health, does this tree appear to need more water, less water, or is it adequately watered?

Please provide a detailed explanation for each answer, referencing visual cues from the image to support your conclusions.
"""

messages = [
    {
        "role": "user",
        "content": [
            {
                "type": "image",
                "image": image,
            },
            {"type": "text", "text": TREE_PROMPT},
        ],
    }
]

# Preparation for inference
text = processor.apply_chat_template(
    messages, tokenize=False, add_generation_prompt=True
)
image_inputs, video_inputs = process_vision_info(messages)
inputs = processor(
    text=[text],
    images=image_inputs,
    videos=video_inputs,
    padding=True,
    return_tensors="pt",
)
inputs = inputs.to("cuda")
# Inference: Generation of the output
generated_ids = model.generate(**inputs, max_new_tokens=512)
generated_ids_trimmed = [
    out_ids[len(in_ids) :] for in_ids, out_ids in zip(inputs.input_ids, generated_ids)
]
output_text = processor.batch_decode(
    generated_ids_trimmed, skip_special_tokens=True, clean_up_tokenization_spaces=False
)
print(output_text[0])
```

## **Results for tree analysis:**
1) **Tree Identification:**
The tree in the image is a **pear tree**. This identification is based on the shape and size of the leaves, as well as the fruit that is hanging from the branches. Pear trees typically have broad, flat leaves and produce round, pear-shaped fruits.

2) **Harvest Information:**
Based on the tree type and the visible maturity of the fruit, the approximate harvest time is likely within the next few weeks. Pears are typically harvested when they are fully colored and have reached their optimal size. The fruits in the image appear to be yellow-green, which is a common indication that they are ripe and ready for picking.

3) **Health Status:**
The tree does not show any visible signs of disease or distress. The leaves are green and healthy-looking, with no discoloration, spots, or abnormal growth. The branches are sturdy, and the overall structure of the tree appears strong and well-maintained.

4) **Water Needs:**
The tree appears to be adequately watered. The leaves are lush and green, indicating that the tree is receiving sufficient moisture. There are no signs of wilting or yellowing leaves, which would suggest that the tree is not getting enough water. The soil around the base of the tree looks moist, further supporting the conclusion that the tree is well-watered.

In summary, the pear tree in the image is healthy, ripe, and ready for harvest. It does not show any signs of disease or distress, and it appears to be adequately watered based on the condition of the leaves and soil.

## **Leaf Disease Analysis**
pear-leaves-with-rust-damage.jpg


```python
image = Image.open("/content/pear-leaves-with-rust-damage.jpg")
# Resize the image, for example, to half its original size
image = image.resize((image.width // 2, image.height // 2))
DISEASE_PROMPT = """Analyze the provided image of the tree's leaves and answer the following questions:

1) Disease Identification: Do the leaves show any visible signs of disease (e.g., discoloration, spots, wilting, or deformities)? If yes, what is the name of the disease or likely condition affecting the leaves?
2) Cause of the Disease: What is the likely cause of this disease (e.g., fungal infection, bacterial infection, pest infestation, or nutrient deficiency)?
3) Cure and Prevention: What are the recommended steps to cure this disease and prevent it from recurring (e.g., specific treatments, pesticides, pruning, or changes in watering or nutrient management)?

Please provide a detailed explanation for each response, citing visual evidence from the image to justify your conclusions."""


messages = [
    {
        "role": "user",
        "content": [
            {
                "type": "image",
                "image": image,
            },
            {"type": "text", "text": DISEASE_PROMPT},
        ],
    }
]

# Preparation for inference
text = processor.apply_chat_template(
    messages, tokenize=False, add_generation_prompt=True
)
image_inputs, video_inputs = process_vision_info(messages)
inputs = processor(
    text=[text],
    images=image_inputs,
    videos=video_inputs,
    padding=True,
    return_tensors="pt",
)
inputs = inputs.to("cuda")
# Inference: Generation of the output
generated_ids = model.generate(**inputs, max_new_tokens=512)
generated_ids_trimmed = [
    out_ids[len(in_ids) :] for in_ids, out_ids in zip(inputs.input_ids, generated_ids)
]
output_text = processor.batch_decode(
    generated_ids_trimmed, skip_special_tokens=True, clean_up_tokenization_spaces=False
)
print(output_text[0])
```

## **Results for leaf disease analysis**
1) Disease Identification:
The leaves in the image show visible signs of disease. The leaves have circular, orange-brown spots that are indicative of a fungal infection known as leaf spot. This disease is characterized by the presence of distinct, circular lesions on the leaves, often with a dark border.

2) Cause of the Disease:
The likely cause of this disease is a fungal infection. The orange-brown spots on the leaves are typical symptoms of fungal leaf spot. Fungi are microscopic organisms that can cause various diseases in plants, and leaf spot is one of the common fungal diseases affecting leaves.

3) Cure and Prevention:
To cure and prevent the recurrence of this disease, the following steps are recommended:

- **Cure:**
  - Remove and dispose of infected leaves: Cut off and remove the leaves with the orange-brown spots and dispose of them to prevent the spread of the fungus.
  - Apply fungicides: Use a fungicide specifically formulated for leaf spot to treat the affected plants. Follow the manufacturer's instructions for proper application and dosage.
  - Improve air circulation: Ensure good air circulation around the plants to reduce the humidity that favors fungal growth.

- **Prevention:**
  - Prune and thin the tree: Remove any dead or damaged branches to improve air circulation and reduce the risk of fungal infections.
  - Maintain proper watering: Avoid overwatering and ensure the soil is well-draining to prevent the buildup of moisture that can promote fungal growth.
  - Use resistant varieties: If possible, plant varieties that are resistant to leaf spot to minimize the risk of infection.
  - Rotate crops: If the tree is part of a larger garden, rotate the planting location to avoid re-infecting the same area with the same pathogen.

By following these steps, the disease can be managed and the risk of recurrence can be minimized.
